```python
from keras.layers import Dense,Conv2D,Flatten,MaxPool2D
from keras.preprocessing.image import ImageDataGenerator
from keras.models import Sequential
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.preprocessing.image import img_to_array,load_img
import numpy as np
import pandas as pd
```

```python
train_csv=pd.read_csv("../input/human-action-recognition-har-dataset/Human Action Recognition/Training_set.csv")
test_csv=pd.read_csv("../input/human-action-recognition-har-dataset/Human Action Recognition/Testing_set.csv")
```

```python
train_csv
```

```python
train_csv["label"].unique()
```

```python
img_data=[]
img_label=[]

for i in range(len(train_csv)):
    img="../input/human-action-recognition-har-dataset/Human Action Recognition/train/"+train_csv["filename"][i]
    img=load_img(img,target_size=(150,150))
    img=img_to_array(img)
    img_data.append(img)
    img_label.append(train_csv["label"][i])
```

```python
img_data=np.array(img_data)
```

```python
img_data.shape
```

```python
img_label=np.array(img_label)
```

```python
img_label.shape
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
img_label=encoder.fit_transform(img_label)
```

```python
img_label=to_categorical(img_label)
```

```python
img_label[0]
```

```python
cnn=Sequential()
cnn.add(Conv2D(filters=32,kernel_size=3,activation="relu",input_shape=(150,150,3)))
cnn.add(MaxPool2D(pool_size=2,strides=2))
cnn.add(Conv2D(filters=32,kernel_size=3,activation="relu"))
cnn.add(MaxPool2D(pool_size=2,strides=2))
cnn.add(Flatten())
```

```python
cnn.add(Dense(units=512,activation="relu"))
cnn.add(Dense(units=15,activation="sigmoid"))
```

```python
cnn.compile(optimizer="adam",loss="categorical_crossentropy",metrics=["accuracy"])
```

```python
cnn.fit(img_data,img_label,epochs=60)
```

```python
cnn.save("human_action.h5")
```
